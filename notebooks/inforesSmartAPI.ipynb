```python
url= 'https://raw.githubusercontent.com/bioregistry/bioregistry/main/src/bioregistry/data/bioregistry.json'
response = requests.get(url)
data = []
count = 0
def make_row(api, data, prefix):
    homepage = api.get('homepage')
    name = api.get('name')
    description = api.get('description')
    row = {
                "short name": prefix,
                "homepage": homepage,
                "name": name,
                "description": description
    }
    data.append(row)
    return data
    
for api in response.json():
    count = count + 1
    prefix_collection = response.json().get(api)
    miriam = prefix_collection.get('miriam')
    obofoundary = prefix_collection.get('obofoundry')
    ols = prefix_collection.get('ols')
    n2t = prefix_collection.get('n2t')
    prefixcommons = prefix_collection.get('prefixcommons')
    if miriam:
        data = make_row(miriam, data, api)
    elif obofoundary:
        data = make_row(obofoundary, data, api)
    elif n2t:
        data = make_row(n2t, data, api)
    else:
        print("trusted resource not found for: " + api)
        
print('total number of resources %', count)
df_description = pd.DataFrame (data, columns = ['short name',
                                     'homepage',
                                     'name',
                                     'description'
                                    ])
#df.to_csv(r'apis.csv', index = False)
df_description.to_csv(r'miriam_registry_prefixes.csv', index = False)
    
```

```python
import json
import requests
from pprint import pprint
import pandas as pd
import bioregistry as br
data = []
registry = bioregistry.read_registry()
for registry_entry in registry:
    homepage = bioregistry.get_homepage(registry_entry)
    name = bioregistry.get_name(registry_entry)
    description = bioregistry.get_description(registry_entry)
    version = bioregistry.get_version(registry_entry)
    if version:
        infores = "infores:"+registry_entry+"."+version
    else:
        infores = "infores:"+registry_entry
    row = {
        "short name": registry_entry,
        "homepage": homepage,
        "name": name,
        "description": description,
        "version": version,
        "infores": infores
    }
    data.append(row)
    
df_description = pd.DataFrame (data, columns = ['short name',
                                     'homepage',
                                     'name',
                                     'description',
                                     'version',
                                     'infores'
                                    ])
#df.to_csv(r'apis.csv', index = False)
df_description.to_csv(r'miriam_registry_prefixes.csv', index = False)
```
