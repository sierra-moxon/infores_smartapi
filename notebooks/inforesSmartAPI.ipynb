```python
import json
import requests
from pprint import pprint
import pandas as pd
import bioregistry as br
import csv
```

```python
reusabledata_url = 'https://raw.githubusercontent.com/sierra-moxon/reusabledata/master/data.json'
sierra_url = 'https://raw.githubusercontent.com/sierra-moxon/reusabledata/add_infores/data-sources/compiled.json'
```

```python
response = requests.get(reusabledata_url)
sierra_response = requests.get(sierra_url)
source_list = []
source_dict = []
infores_list = []
mapped_licenses = []

csv_infores = 'https://raw.githubusercontent.com/sierra-moxon/reusabledata/add_infores/infores.csv'

for source in response.json():
    source_list.append(source.get('source'))
    bits = {"id":source.get('source'),"license":source.get('license'), "license_type":source.get('license-type'),
            "license_commentary": source.get('license-commentary')}
    source_dict.append(bits)
    
for ssource in sierra_response.json():
    source_list.append(ssource.get('id'))
    bits = {"id":ssource.get('id'),"license":ssource.get('license'), "license_type":ssource.get('license-type'),
            "license_commentary": ssource.get('license-commentary')}
    source_dict.append(bits)
csv_infores_response = requests.get(csv_infores)
with requests.Session() as s:
    download = s.get(csv_infores)

    decoded_content = download.content.decode('utf-8')

    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
    for row in my_list:
        infores_list.append(row[5])
        infores_list.append(row[6].replace("infores:",""))

intersection = set(source_list).intersection(infores_list)
for item in intersection:
    for mapper in source_dict:
        if mapper.get("id") == item:
            mapped_licenses.append(mapper)

keys = mapped_licenses[0].keys()
with open('/tmp/mapped_licenses.csv', 'w', newline='')  as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(mapped_licenses)
            
```
