```python
import json
import requests
from pprint import pprint
import pandas as pd
```

```python
url= 'https://smart-api.info/api/query?q=tags.name:translator&size=150&fields=info,paths,servers,tags,components.x-bte*,info,_meta'
response = requests.get(url)
data = []

for api in response.json().get('hits'):
    info = api.get('info')
    
    xtranslator = info.get('x-translator')
    if xtranslator is not None:
        component = xtranslator.get('component')
        team = xtranslator.get('team')
        biolink_version = xtranslator.get('biolink-version')
        team = xtranslator.get('team')
    else:
        component = ""
        team = ""
        biolink_version = ""
        team = ""
    
    xtrapi = info.get('x-trapi')
    if xtrapi is not None:
        trapi_version = xtrapi.get('version')
    else:
        trapi_version = 'unknown'
    title = info.get('title')
    
    description = info.get('description')
    contact = info.get('contact')
    contact_name = contact.get('name')
    print(contact_name)
    contact_email = contact.get('email')
    servers = api.get('servers')
    if servers is not None:
        for server in servers:
            server_url = server.get('url')
            server_description = server.get('description')
            row = {
                "title": title,
                "component": component,
                "team":team,
                "trapi_version":trapi_version,
                "biolink_version":biolink_version,
                "contact_name":contact_name,
                "contact_email":contact_email,
                "server_url": server_url,
                "server_description": server_description
            }
            data.append(row)
    else:
        server_url = "unknonwn"
        server_description = "unknown"

        row = {
                "title": title,
                "component": component,
                "team":team,
                "trapi_version":trapi_version,
                "biolink_version":biolink_version,
                "contact_name":contact_name,
                "contact_email":contact_email,
                "server_url": server_url,
                "server_description": server_description
        }
        data.append(row)
    
df = pd.DataFrame (data, columns = ['title','component','team','trapi_version','biolink_version','contact_name', 'contact_email', 'server_url', 'server_description'])
                                    #,'team','trapi_version','biolink_version','contact_name','contact_email'])
df.to_csv(r'apis.csv', index = False)
# print(df)

    
```
