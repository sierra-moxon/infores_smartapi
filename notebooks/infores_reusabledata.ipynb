```python
import json
import requests
from pprint import pprint
import pandas as pd
import bioregistry as br
import csv
import bmt

tk = bmt.Toolkit()
url= 'https://smart-api.info/api/query?q=tags.name:translator&size=150&fields=info,servers,paths'
response = requests.get(url)
data = []
count = 0
for api in response.json().get('hits'):
    info = api.get('info')
    paths = api.get('paths') 
    title = info.get('title')
    api_id = api.get('_id')
    servers = api.get('servers')
    if servers is not None:
        for server in servers:
            for path in paths:
                if path == '/meta_knowledge_graph':
                    if server.get('url')+path == '/jan2021/meta_knowledge_graph' or server.get('url')+path == '--proxy-headers/meta_knowledge_graph':
                        continue
                    else:
                #if path == '/predicates':
                        count = count + 1
                        data.append(server.get('url')+path)
print(count)
predicates = []
for meta_graph in data:
    if meta_graph in ('https://unsecret.ncats.io/v2/meta_knowledge_graph',
                      'https://explanatory-agent.azurewebsites.net/v1.1//meta_knowledge_graph',
                      'https://arax.ncats.io/api/rtxkg2/v1.1/meta_knowledge_graph'):
        continue
    else:
        response = requests.get(meta_graph)
        if response.status_code != 404 and response.status_code != 502:
            for edge in response.json().get('edges'):
                subject = edge.get('subject')
                predicate = edge.get('predicate')
                object = edge.get('object')
                edge = (subject, predicate, object)
                if edge not in predicates:
                    predicates.append(edge)

f = open('/tmp/predicates.txt', 'w')

# create the csv writer
writer = csv.writer(f)
                                        
for predicate in predicates:
    if 'increase' in predicate[1] or 'decrease' in predicate[1]:
        writer.writerow([predicate[0],predicate[1],predicate[2]])
            
# close the file
f.close()
```
